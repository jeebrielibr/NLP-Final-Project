## Normalisasi Dataset

```python
import pandas as pd
import re

df1 = pd.read_csv('../Dataset/datasetraw1.csv')
df2 = pd.read_csv('../Dataset/datasetraw2.csv')
df3 = pd.read_csv('../Dataset/datasetraw3.csv')
df4 = pd.read_csv('../Dataset/datasetraw4.csv')
df5 = pd.read_csv('../Dataset/datasetraw5.csv')
df6 = pd.read_csv('../Dataset/datasetraw6.csv')

df1_std = df1[['komentar', 'label']].rename(columns={'komentar': 'text', 'label': 'label'})
df2_std = df2[['message', 'label']].rename(columns={'message': 'text', 'label': 'label'})
df3_std = df3[['text', 'label']].rename(columns={'text': 'text', 'label': 'label'})
df4_std = df4[['comment', 'label']].rename(columns={'comment': 'text', 'label': 'label'})
df5_std = df5[['text', 'label']].rename(columns={'text': 'text', 'label': 'label'})
df6_std = df6[['comment', 'label_id']].rename(columns={'comment': 'text', 'label_id': 'label'})

print("Dataset 1:", df1_std.head())
print("Dataset 2:", df2_std.head())
print("Dataset 3:", df3_std.head())
print("Dataset 4:", df4_std.head())
print("Dataset 5:", df5_std.head())
print("Dataset 6:", df6_std.head())
```

## Gabungkan Dataset

```python
df_combined = pd.concat([df1_std, df2_std, df3_std, df4_std, df5_std, df6_std], ignore_index=True)
print(f"Total data awal: {len(df_combined)}")
```

## Handling Missing Values

```python
df_combined = df_combined.dropna(subset=['text', 'label'])
```

## Penanganan Duplikat

```python
df_combined = df_combined.drop_duplicates(subset=['text'])
print(f"Total data setelah hapus duplikat: {len(df_combined)}")
```

## Pembersihan Dasar

```python
def basic_clean(text):
    if not isinstance(text, str):
        return ""
    # Lowercase
    text = text.lower()
    # Hapus whitespace berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_combined['text'] = df_combined['text'].apply(basic_clean)
df_combined.head()
```

## Validasi Label

```python
# Memastikan label adalah integer 0 atau 1
df_combined['label'] = df_combined['label'].astype(int)
df_combined = df_combined[df_combined['label'].isin([0, 1])]
```

## Cek Distribusi Label

```python
print("\nDistribusi Label:")
print(df_combined['label'].value_counts())
```

## Simpan Dataset

```python
output_path = '../Dataset/dataset_prepared.csv'
df_combined.to_csv(output_path, index=False)
print(f"\nDataset siap disimpan di: {output_path}")
```
